### Cargar y listar fichero XML

```python
from lxml import etree

fichero = input("Introduzca nombre fichero: ")
try: 
    doc = etree.parse(fichero)
    print(etree.tostring(doc, pretty_print=True, xml_declaration=True, encoding="utf-8").decode())
except:
    print("Fichero no encontrado")
```

### Filtrar fichero XML con Xpath

```python
expresion = input("Introduzca expresión XPath: ")
lista_resultado = doc.xpath(expresion)
for item in lista_resultado:
    print(item)
```

### Transformar con XSL

```python
doc_xml = etree.parse('biblioteca.xml')
doc_xsl = etree.parse('biblioteca.xsl')
transform = etree.XSLT(doc_xsl)
resultado = transform(doc_xml)
print(str(resultado))
```
